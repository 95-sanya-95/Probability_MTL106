**Problem statement** : Sanya (2022MT11286) and Alice are managing teams of warriors. Alice's team has warriors with strengths 3, 4, 6, and 7. In each match, the warrior with strength x wins with a probability of x/(x+y), and their strength becomes x+y after the match. Assuming optimal play, determine the probability of you winning the tournament

***Initial situation:***

Alice = [3,4,6,7] as the strength of warriors

Sanya = [1,2,8,6] as the strength of warriors { My entry number is 2022MT1**1286** }

Sanya always picks the best warrior for each fight

```python
import random
import math
import matplotlib.pyplot as plt
%matplotlib inline
import time

start_time = time.time()

def best_warrior(iterations):
    counter = 0 # iterated for the number of iterations
    no_of_wins = 0 # calculate the number of wins of Sanya in the number of iterations done
    win_percentages = []
    while(counter<iterations):
        alice = [3,4,6,7]
        sanya = [1,2,8,6]  # my entry number is 2022MT11286
        while len(sanya) > 0 and len(alice) > 0:

            warrior_alice = random.choice(alice) # this funciton will give the choice of alice, its a random function therefore the probability of choice of each is same

            warrior_sanya = max(sanya) # this will give the maximum strength warrior from the array sanya

            win = ["sanya","alice"]
            probabilities  = [(warrior_sanya /(warrior_sanya  + warrior_alice)),(warrior_alice/(warrior_sanya  + warrior_alice))] # here I have given the probabilities of each warrior winning
            who_win = random.choices(win, weights=probabilities)[0] # a simple random fuunction chooses each element with equal probability,this random function will choose the element according to the probabilitie given in the above array
            if(who_win=="sanya"):
                alice.remove(warrior_alice) # when sanya wins then acc. to the rule the warrior of the alice dies so I have removed the warrior of alice and added the strength of theat warrior to the warrior of Sanya
                sanya.remove(warrior_sanya)
                sanya.append(warrior_alice+warrior_sanya)

            else:
                alice.remove(warrior_alice) # when alice wins then acc. to the rule the warrior of the sanya dies so I have removed the warrior of sanya and added the strength of theat warrior to the warrior of alice
                sanya.remove(warrior_sanya)
                alice.append(warrior_alice+warrior_sanya)

         # after reaching here either of sanya or alice warrior array would have become empty

        if(len(sanya)>0): # if sanya wud have warriors left (length of sanya array is more than 0) then she will be the winner
            no_of_wins+=1
        counter+=1

        win_percentages.append(no_of_wins / counter) # I am using this to keep the track of the wins sanya has until that number of iterations to plot the graph

    plt.plot(range(1, iterations + 1), win_percentages, label='Win Percentage')
    plt.axhline(y=0.5, color='r', linestyle='--', label='50% Chance')
    plt.xlabel('Number of Simulations')
    plt.ylabel('Win Percentage of Sanya')
    plt.title('Probability of Winning of Sanya vs. Number of Simulations')
    plt.legend()
    plt.show()

    ans = no_of_wins / iterations # monte-carlo probability for sanya winning
    return ans

answer_best = best_warrior(100000) # no. of iterations are 10^5

end_time = time.time()

print("                    ")
print("                    ")
print("                    ")
print("When Sanya chooses best warrior, then the probabilty after 10^5 iteraitions is", answer_best)

execution_time = end_time - start_time
print("                    ")
print("                    ")
print("                    ")
print(f"Execution time: {execution_time} seconds")
```

If Alice picks a warrior with strength x and I pick a warrior with strength y, what is the expected gain of strength in your team ? Is every match fair(i.e. 0 expected gain) ?

**Elaboration on Expected Gain Calculation**

Let's denote:

- $x$: Alice's chosen value.
- $y$: Sanya's chosen value.
- $X$: A random variable representing the gain. It takes the value $x$ if Sanya wins and $-y$ if Sanya loses.


1. **When Sanya wins**: The probability of Sanya winning is $\frac{y}{x+y}$ (because her winning probability is her warrior strength divided by the total warrior strength). In this case, Alice loses her warrior strength $x$ and Sanya gains it, so the gain is $x$.

2. **When Sanya loses**: The probability of Sanya losing is $\frac{x}{x+y}$ (because her losing probability is her warrior strength divided by the total warrior strength). In this case, Alice gains Sanya's warrior strength $y$ (as a negative value) and Sanya loses it, so the gain is $-y$.

Now, we can calculate the expected gain by taking the weighted average of the gains in these two scenarios:


$$
E(\text{gain}) = (x \cdot \frac{y}{x+y}) + (-y \cdot \frac{x}{x+y}) \\
= 0
$$

So, this means that the expected value of gain is 0.

**And this means that it does not matter how Sanya plays the expected gain is always zero. Therefore in the task 4, optimal approach does not exist. So, I have just choosen the random warrior for Sanya.**

## **Theoretical probability**

**Claim:** Let $prob(n)$ be the probability that Sanya wins a match among $n$ warriors, where $n = p_{n} + q_{n}$, and $p_{n}$ and $q_{n}$ are the number of warriors of Sanya and Alice respectively.

**Proof:**

**Base condition:** When the total number of warriors is $1$, either Sanya wins (if she has a warrior) or Alice wins. Therefore, the probability is either $0$ or $1$, depending on whether Sanya has a warrior.

**Induction hypothesis:** Assume that for $n-1$ warriors, the probability $prob(n-1)$ is given by where $n-1 = p_{(n-1)} + q_{(n-1)}$
$$
prob(n-1) = \frac{s_1 + s_2 + s_3 + \dots + s_{p_{(n-1)}}}{(s_1 + s_2 + s_3 + \dots + s_{p_{(n-1)}}) + (a_1 + a_2 + a_3 + \dots + a_{q_{(n-1)}})}
$$

**Induction step:** We consider two cases: Sanya wins the first match or Alice wins.

1. **Sanya wins the first match:**
$$
\text{Probability} = \frac{s_1}{s_1 + a_1} \times prob(n-1)
$$

2. **Alice wins the first match:**
$$
\text{Probability} = \frac{a_1}{s_1 + a_1} \times prob(n-1)
$$

Now, let's compute the probability using these two cases:

$$

\text{prob(n)}  = \frac{s_1}{s_1 + a_1} \times prob(n-1) + \frac{a_1}{s_1 + a_1} \times prob(n-1) \\
= \frac{s_1(a_1 +s_1+ s_2 + s_3 + \dots + s_{p_{n}}) + a_1(s_2 + s_3 + \dots + s_{p_n})}{(s_1 + a_1)(s_1 + s_2 + s_3 + \dots + s_{p_n} + a_1 + a_2 + a_3 + \dots + a_{q_n})} \\
= \frac{s_1(s_1 + s_2 + s_3 + \dots + s_{p_n}) + a_1(s_1 + s_2 + s_3 + \dots + s_{p_n})}{(s_1 + a_1)(s_1 + s_2 + s_3 + \dots + s_{p_n} + a_1 + a_2 + a_3 + \dots + a_{q_n})} \\
= \frac{(s_1 + a_1)(s_1 + s_2 + s_3 + \dots + s_{p_n})}{(s_1 + a_1)(s_1 + s_2 + s_3 + \dots + s_p + a_1 + a_2 + a_3 + \dots + a_{q_n})} \\
= \frac{s_1 + s_2 + s_3 + \dots + s_{p_n}}{(s_1 + s_2 + s_3 + \dots + s_{p_n}) + (a_1 + a_2 + a_3 + \dots + a_{q_n})}

$$

Hence, the claim is proved by induction.
